```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
g1 = nx.erdos_renyi_graph(20,0.5)
nx.draw(g1)
plt.show()
```

```python
l = []
for i in range(20):
    for neb in g1.neighbors(i):
        l = g1.degree(neb)
        #print (l)
    print (g1.degree(i), np.mean(l))
```
